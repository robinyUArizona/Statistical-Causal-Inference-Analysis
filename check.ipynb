```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Example scores and true labels
np.random.seed(0)  # For reproducibility
scores = np.random.rand(1000)  # Replace with your model scores
true_labels = np.random.randint(0, 2, 1000)  # Replace with your true labels

# Step 1: Create histogram of model scores
plt.figure(figsize=(12, 5))

# Histogram of model scores
plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
plt.hist(scores[true_labels == 1], bins=30, alpha=0.5, label='Class 1 (Fraud)', color='red')
plt.hist(scores[true_labels == 0], bins=30, alpha=0.5, label='Class 0 (Non-fraud)', color='blue')
plt.xlabel('Model Score')
plt.ylabel('Frequency')
plt.legend()
plt.title('Histogram of Model Scores by Class')

# Step 2: Define thresholds
thresholds = np.linspace(0, 1, 100)

# Step 3: Calculate FPR for each threshold
fpr = []
for threshold in thresholds:
    predictions = (scores >= threshold).astype(int)
    tn, fp, fn, tp = confusion_matrix(true_labels, predictions).ravel()
    fpr.append(fp / (fp + tn))

# Step 4: Plot FPR vs. Threshold
plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
plt.plot(thresholds, fpr, label='FPR', color='orange')
plt.axhline(y=0.05, color='r', linestyle='--', label='Target FPR = 5%')
plt.xlabel('Threshold')
plt.ylabel('False Positive Rate')
plt.title('FPR vs. Threshold')
plt.legend()

# Show the plots
plt.tight_layout()
plt.show()

# Step 5: Find the threshold corresponding to FPR close to 5%
target_fpr = 0.05
closest_index = (np.abs(np.array(fpr) - target_fpr)).argmin()
optimal_threshold = thresholds[closest_index]

print(f'Optimal Threshold for FPR close to 5%: {optimal_threshold}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Example scores and true labels
np.random.seed(0)  # For reproducibility
scores = np.random.rand(1000)  # Replace with your model scores
true_labels = np.random.randint(0, 2, 1000)  # Replace with your true labels

# Step 1: Define thresholds
thresholds = np.linspace(0, 1, 100)

# Step 2: Calculate the frequency of scores above each threshold
frequencies = []
for threshold in thresholds:
    # Count scores above the threshold
    count_above_threshold = np.sum(scores >= threshold)
    frequencies.append(count_above_threshold)

# Step 3: Plot the histogram of frequencies against thresholds
plt.figure(figsize=(10, 5))
plt.bar(thresholds, frequencies, width=0.01, color='lightblue', edgecolor='black')
plt.axhline(y=0.05 * len(scores), color='r', linestyle='--', label='Target Frequency for 5% FPR')
plt.xlabel('Threshold')
plt.ylabel('Frequency of Scores Above Threshold')
plt.title('Frequency of Model Scores Above Threshold')
plt.legend()
plt.grid()

# Show the plot
plt.show()

# Step 4: Find the threshold corresponding to the desired frequency
target_frequency = 0.05 * len(scores)  # 5% of total scores
closest_index = (np.abs(np.array(frequencies) - target_frequency)).argmin()
optimal_threshold = thresholds[closest_index]

print(f'Optimal Threshold for FPR close to 5%: {optimal_threshold}')
```
